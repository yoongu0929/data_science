```python
lat_lng = pd.read_csv("latitude_longitude_final_190610.csv")
lat_lng.drop(columns=['Unnamed: 0'] , axis=1, inplace=True)
lat_lng
```

```python
# 116199개
wine_data_js = pd.read_csv("wine_raw_data_js.csv")
wine_data_js
```

```python
def make_name(row):
    if row['region_1'] is np.nan:
        return ", ".join([row['country'], row['province']])
    return ", ".join(set([row['country'], row['province'], row['region_1']]))
    
wine_data_js["name"] = wine_data_js.apply(lambda row: make_name(row), axis=1)
```

```python
wine_data_js
```

```python
# 와인데이터 + 위도경도 merge
wine_data_lat_lng_merge = pd.merge(wine_data_js, lat_lng, how='left', on=['country', 'province', 'region_1'])
wine_data_lat_lng_merge
```

```python
# 와인데이터 + 위도경도 merge 후 컬럼 정리
wine_data_lat_lng_merge.drop(columns=['Unnamed: 0', 'name_x', 'name_y', 'location'] , axis=1, inplace=True)
wine_data_lat_lng_merge
```

```python
# 위도/경도 소수점 자리 조정
wine_data_lat_lng_merge['longitude'] = round(wine_data_lat_lng_merge['longitude'],1)
wine_data_lat_lng_merge['latitude'] = round(wine_data_lat_lng_merge['latitude'],1)
wine_data_lat_lng_merge
```
