```python
wine_ohe = pd.get_dummies(wine, prefix=['main_country', 'main_variety'])
one_hot1 = wine_ohe.drop(columns= ["Unnamed: 0", "taster_name", "price", "stand_points", "points"], axis=1)
one_hot2 = wine_ohe.drop(columns= ["Unnamed: 0", "taster_name", "price", "norm_points", "points"], axis=1)
one_hot3 = wine_ohe.drop(columns= ["Unnamed: 0", "taster_name", "price", "norm_points", "stand_points"], axis=1)
one_hot4 = wine_ohe.drop(columns= ["Unnamed: 0", "taster_name", "norm_points", "stand_points", "points"], axis=1)
one_hot4.head()
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#t Linear Regression
def linear_reg(dm_x, dm_y=wine[["price"]]):
    x_train, x_test, y_train, y_test = train_test_split(dm_x, dm_y, test_size=0.3, random_state=1)
    reg = linear_model.LinearRegression()
    reg.fit(x_train, y_train)
    predict_result = reg.predict(x_test)

    print("accuracy: "+ str(reg.score(x_test, y_test)*100) + "%")
    print("Mean absolute error: {}".format(mean_absolute_error(y_test, predict_result)))
    print("Mean squared error: {}".format(mean_squared_error(y_test, predict_result)))
    R2 = r2_score(y_test, predict_result)
    print('R Squared: {}'.format(R2))

    n=x_test.shape[0]
    p=x_test.shape[1] - 1
    adjusted_r_squared = 1 - (1 - R2) * ((n - 1)/(n-p-1))
    print('Adjusted R Squared: {}'.format(adjusted_r_squared))
```

```python
linear_reg(one_hot1)
```

```python
linear_reg(one_hot2)
```

```python
linear_reg(one_hot3)
```

```python
linear_reg(one_hot4, wine[["points"]])
```
