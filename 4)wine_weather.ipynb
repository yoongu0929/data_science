```python
we = pd.read_csv("final_rain_temp_1991-2016.csv")
```

```python
we.drop("Unnamed: 0", axis=1, inplace=True)
we = we.drop_duplicates()
```

```python
we
```

```python
wine_latlng = pd.read_csv("wine_latlng.csv")
wine_latlng = wine_latlng.drop(columns =["Unnamed: 0", "year"])
wine_latlng
```

```python
wine_weather = pd.merge( wine_latlng, we,  on=["latitude","longitude"], how="left")
wine_weather
```

```python
wine_weather.to_csv("wine_weather.csv")
```

```python
wine_pcv = pd.read_csv("wine_df.csv")
wine_pcv
```

```python
weather_points_coutry_variety = pd.merge( wine_pcv, wine_weather,  right_on="Unnamed: 0_x",  left_on="Unnamed: 0")
weather_points_coutry_variety
```

```python
weather_points_coutry_variety = weather_points_coutry_variety.drop(["Unnamed: 0_x", "latitude", "longitude", "Unnamed: 0"], axis=1)
weather_points_coutry_variety['year'] = weather_points_coutry_variety['year'].astype("int")
weather_points_coutry_variety
```

```python
weather_points_coutry_variety.to_csv("wine_data_for_LR.csv")
```
