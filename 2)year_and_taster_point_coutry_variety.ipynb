```python
wine_raw_data = pd.read_csv("title_without_winery.csv")

# 필요없는 컬럼 삭제
wine_raw_data.drop(columns= ["Unnamed: 0","taster_twitter_handle"] , axis=1, inplace=True)
# price랑 country Nan 데이타 제거
wine_raw_data.dropna(subset=['price', 'country'], inplace = True)
wine_raw_data.drop_duplicates(inplace = True)
```

```python
import re

def get_year(data):
    years = re.findall("\d+", data)
    if len(years) > 0:
        year = sorted([int(year) for year in years], reverse=True)[0]
        return year if year <= 2017 and year >= 1998 else 0
    else:
        return 0
    
wine_raw_data["year"] = wine_raw_data["title"].apply(lambda data: get_year(data))
wine_raw_data= wine_raw_data[wine_raw_data["year"] != 0]
```

```python
wine_raw_data.to_csv("year_raw_data.csv")
```

```python
wine1 = wine_raw_data[['country', 'points', 'price', 'taster_name', 'variety', 'year']]
wine1 # 107139 개
```

```python
# taster_name 분류를 위한 문자열 변환
wine1["taster_name"] = wine1["taster_name"].astype('str')

# taster_name 정수형 변환
def mapping(data,feature):
    featureMap=dict()
    count=0
    for i in sorted(data[feature].unique(),reverse=True):
        featureMap[i]=count
        count=count+1
    data[feature]=data[feature].map(featureMap)
    return data.head()

mapping(wine1, "taster_name")
```

```python
# taster별 dataframe 생성
wine_tasters = []
for i in range (max(wine1["taster_name"])+1):
    wine_tasters.append(wine1[wine1["taster_name"] == i ])

    
# 정규화 및 표준화
from sklearn import preprocessing

for wine in wine_tasters:
    wine["norm_points"] = preprocessing.minmax_scale(wine["points"])
    wine["stand_points"] = preprocessing.scale(wine["points"])
    
wine_df = pd.DataFrame(columns=["points", "taster_name"])
while len(wine_tasters) > 0 :
    wine_df = pd.concat([wine_df, wine_tasters[0]])
    del wine_tasters[0]
       
wine_df.sort_index(inplace = True)
wine_df # 107139 개
```

```python
# 상위 1% 나라 외 other로 변경 [main_country]
percent = 0.01
wine_c = wine_df.groupby(["country"]).size().reset_index(name="counts")
wine_c["percent"] = wine_c["counts"]/np.sum(wine_c.counts)
country_list = wine_c[wine_c["percent"] > percent]["country"].unique().tolist()
print(country_list)

wine_df["main_country"] = wine_df["country"].apply(lambda x : x if x in country_list else "other")
wine_df
```

```python
# 동일 포도품종명 변경
variety_to = {
    "Shiraz" : 'Syrah',
    "Fumé Blanc" : "Sauvignon Blanc", 
    "Sauvignon" : "Sauvignon Blanc",
    "Primitivo" : "Zinfandel",
    "Zierfandler" : "Zinfandel",
    "Rosato" : "Rosé"
}
wine_df["variety"] = wine_df["variety"].replace(variety_to)

# 상위 30개  포도품종 외 other로 변경 [main_variety]
wine_v = wine_df.groupby(["variety"]).size().reset_index(name="counts").sort_values(by="counts", ascending=False)[:30]
variety_list = wine_v["variety"].unique().tolist()
print(variety_list)
wine_df["main_variety"] = wine_df["variety"].apply(lambda x : x if x in variety_list else "other")
wine_df #107139 row
```
