```python
import re

wine_raw_data = pd.read_csv("title_without_winery.csv")

# 필요없는 컬럼 삭제
wine_raw_data.drop(columns=['Unnamed: 0','taster_twitter_handle'] , axis=1, inplace=True)

# price랑 country Nan 데이타 제거
wine_raw_data.dropna(subset=['price', 'country'], inplace = True)
wine_raw_data.head(5)
```

```python
wine0 = wine_raw_data[["price", "country", "points", "variety"]]
wine0.head()
```

```python
import statsmodels.api as sm
from patsy import *
```

```python
model1 = sm.OLS.from_formula("price~ country", wine0)
print(model1.fit().summary())
```

```python
wine1 = pd.read_csv("wine_pcv.csv")
wine1.head()
```

```python
model2 = sm.OLS.from_formula("price~ main_country", wine1)
print(model2.fit().summary())
```
