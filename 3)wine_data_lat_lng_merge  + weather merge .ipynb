```python
# 107139 개
wine_data = pd.read_csv("year_raw_data.csv")
wine_data
```

```python
def make_name(row):
    if row['region_1'] is np.nan:
        return ", ".join([row['country'], row['province']])
    return ", ".join(set([row['country'], row['province'], row['region_1']]))
    
wine_data["name"] = wine_data.apply(lambda row: make_name(row), axis=1)
```

```python
wine_data
```

```python
# 와인데이터 + 위도경도 merge
wine_data_lat_lng_merge = pd.merge(wine_data, lat_lng, how='left', on=['country', 'province', 'region_1'])
wine_data_lat_lng_merge
```

```python
# 와인데이터 + 위도경도 merge 후 컬럼 정리
wine_latlng = wine_data_lat_lng_merge[["Unnamed: 0_x", "year", "latitude", "longitude"]]
wine_latlng
```

```python
wine_latlng.to_csv("wine_latlng.csv")
```
